```python
# randn = 표준 정규분포(standard normal distribution)에서 난수 생성
# 정규분포 : 평균을 중심으로 좌우 대칭인 종 모양의 분포. 가운데 평균값이 가장 많이 몰림. 완전히 대칭
# 평균(mean) = 0   표준편차(std) = 1

df4 = pd.DataFrame(np.random.randn(10, 4))
df4
```

```python
pieces = [df4[:3], df4[3:7], df4[7:]]
pd.concat(pieces)
```
