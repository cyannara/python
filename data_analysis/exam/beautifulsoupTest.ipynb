```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.naver.com/"
response = requests.get(url)

if response.status_code == 200:
  html = response.text
  print("웹 페이지 소스 가져오기 성공!")
  soup = BeautifulSoup(html, 'html.parser')
  title_tag = soup.title
  print("제목 태그:" , title_tag)
  print("제목 텍스트:", title_tag.get_text())

  # id가 query인 input 태그 찾기
  search_box = soup.find('input', id='query')
  print('검색창의 name 속성', search_box.get('name'))
  print("-" *20)
else:
  print(f"웹 페이지 소스 가져오기 실패. 상태코드: { response.status_code}")
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
```

```python
driver = webdriver.Chrome()
driver.get("https://www.google.com")
time.sleep(2)

print(f"현태 페이지 제목: {driver.title}")

#검색창 요소 찾기
search_box = driver.find_element(By.NAME,'q')

#검색창에 '셀레늄' 입력
search_box.send_keys('셀레늄')
time.sleep(3)

#검색버튼 클릭
search_button = driver.find_element(By.NAME, 'btnK')
search_button.click()
time.sleep(5)

driver.quit()
```
