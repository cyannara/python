```python
from faker import Faker
import pandas as pd 
import numpy as np
import random
```

```python
# Faker 객체 생성
# 한국어 데이터를 생성하려며 'ko_kr' 로케일을 지정
fake = Faker("ko_kr")

#다양한 가짜 데이터 생성
print("이름:", fake.name())
print("주소:", fake.address())
print("상세주소:", fake.address_detail())
print("전화번호:", fake.phone_number())
print("이메일:", fake.email())
print("직업:", fake.job())
print("회사명:", fake.company())
print("국가:", fake.country())
print("성별:", random.choice(['남','여']) )
```

```python
# Faker로 대랴의 데이터 만들기
data = []

for _ in range(100):
  name= fake.name()
  age = np.random.randint(20,30)
  gender = np.random.choice(['남','여'])
  score = np.random.randint(50,101)
  data.append([name, age, gender, score])

# dataframe 생성
columns = ['이름', '나이','성별','점수']
df = pd.DataFrame(data, columns=columns)

print("생성된 가짜 데이터프레임 (상위 5개 행):")
print(df.head())
```

```python
data = []

for _ in range(100):
  name= fake.name()
  dept = np.random.choice(['개발팀','영업팀','마케팅팀'])
  position = np.random.choice(['사원','대리','과장'])
  
  data.append([name, dept, position])

# dataframe 생성
columns = ['이름', '부서','직급']
df = pd.DataFrame(data, columns=columns)

print("생성된 가짜 데이터프레임 (상위 5개 행):")
print(df.head())

df.to_csv('company.csv')
```
