```python
import folium
```

```python
# 서울의 중심 좌표 (위도, 경도)
seoul_coords = [37.5665, 126.9780]

seoul_map = folium.Map(location=seoul_coords, zoom_start=12)

# 서울의 랜드마크 위치 데이터
landmarks = {
    '경복궁': [37.5796, 126.9771],
    '남산타워': [37.5512, 126.9882],
    '롯데월드타워': [37.5126, 127.1026]
}

# 랜드마크 위치에 마커 추가
for name, coords in landmarks.items():
    folium.Marker(
        location=coords,
        popup=name,  # 마커 클릭 시 나타나는 팝업 내용
        tooltip=name, # 마커에 마우스를 올렸을 때 나타나는 텍스트
        icon=folium.Icon(color='red', icon='info-sign')
    ).add_to(seoul_map)
    
seoul_map.save('seoul_map.html')
```

```python
import pandas as pd
# 가상의 매장 위치 데이터프레임 생성
store_data = {
    '매장명': ['강남점', '홍대점', '건대점'],
    '위도': [37.4981, 37.5566, 37.5408],
    '경도': [127.0276, 126.9237, 127.0694],
    '판매량': [150, 120, 95]
}
stores_df = pd.DataFrame(store_data)

# 서울 중심 지도 생성
seoul_map = folium.Map(location=[37.5665, 126.9780], zoom_start=12)

# 데이터프레임의 각 행을 반복하며 마커 추가
for index, row in stores_df.iterrows():
    folium.CircleMarker(
        location=[row['위도'], row['경도']],
        radius=row['판매량'] / 20, # 판매량에 따라 원의 크기 조절
        color='blue',
        fill=True,
        fill_color='blue',
        tooltip=f"{row['매장명']} (판매량: {row['판매량']})"
    ).add_to(seoul_map)

seoul_map.save('seoul_map.html')    
```
