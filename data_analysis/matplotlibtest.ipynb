```python
import matplotlib.pyplot as plt

data = [10,20,30,40,50]

plt.plot(data)

plt.show()
```

```python
x_values = [1,3,4,5,7]
y_values = [10,20,30,40,50]

plt.plot(x_values, y_values)

plt.show()
```

```python
import seaborn as sns
import pandas as pd
import numpy as np

plt.rcParams['font.family'] = 'Malgun Gothic'
```

```python
tips = sns.load_dataset('healthexp')
print('tips 데이터셋 상위 5개 행')
print(tips.head())
print(tips.size)
```

```python
df = sns.load_dataset('geyser')
print(df.head())

# scatter : 흩어지다. 
# scatterplot : 산점도 함수 , 두 숫자 데이터의 관계를 점으로 표현하는 그래프
sns.scatterplot(data=df, x='duration', y='waiting', hue='kind')
```

```python
tips = sns.load_dataset('tips')
print(tips.head())

# scatter : 흩어지다. 
# scatterplot : 산점도 함수 , 두 숫자 데이터의 관계를 점으로 표현하는 그래프
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex')
```

```python
print(tips.columns)
sns.barplot(data=tips, x='smoker', y='tip')
```

```python
sns.boxplot(
    data=tips,
    x="day",
    y="tip"
)
```
