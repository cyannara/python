# 일차원 데이터

```python
import numpy as np
import pandas as pd
```

```python
pd.Series(7, )
```

```python
# 표준정규분포(가우시안분포)를 따르는 난수를 생성
# 평균(mean) = 0
# 표준편차(std) = 1     
# 발생범위  -1~1:68%,  -2~2:95%, -3~3:99.7%  
s = pd.Series(np.random.randn(5))     # -3 ~ 3
s = pd.Series(np.random.randn(5)*10)  # 스케일 조정 (표준편차 확대)
s = pd.Series(np.random.randn()+100)  # 평균이동.  평균 100근처 값
s
```

```python
s = pd.Series(np.random.uniform(0,10,5))
s
```

```python
x1 = [1,3,5,7,9]
x2 = [3,4,5,6,7]
s1 = pd.Series(x1)
s2 = pd.Series(x2)

print(s1.var(ddof=0), s1.std(ddof=0))  # 모집단 표준편차  <- 전체데이터   ddof(자유도):Delta Degrees of Freedom
print(s1.var(), s1.std())              # 표본 표준편차    <- 전체 중 일부만 추출. 실무에서 사용. 표본분산이 더 큼(퍼짐을 보정)
```
